### estimate the time signature of worst tracks from the .beats files

```python
# Simple initial function to detect the time signature of a song

import numpy as np
from collections import Counter

def detect_time_signature(beats_file):
    """Detect the time signature of a song based on the beat annotations."""
    # Load beats data
    beat_times, beat_numbers = np.loadtxt(beats_file, unpack=True)

    # Find repeating cycle pattern
    diffs = np.diff(beat_numbers)  # Differences between consecutive beats
    cycle_lengths = []

    current_cycle = []
    for i in range(len(diffs)):
        current_cycle.append(beat_numbers[i])
        if diffs[i] < 0:  # Detect restart of the beat cycle
            cycle_lengths.append(len(current_cycle))
            current_cycle = []

    # Count occurrences of cycle lengths
    if cycle_lengths:
        most_common_cycle = Counter(cycle_lengths).most_common(1)[0][0]
        return f"{most_common_cycle}/4"
    else:
        return "Unable to determine time signature."

beats_file = "/content/data/annotations/gtzan/annotations/beats/gtzan_blues_00000.beats"
print(detect_time_signature(beats_file))
```

# Analysis of time signatures in the training set

```python
import os
import numpy as np
import json
import matplotlib.pyplot as plt
from collections import defaultdict, Counter

def parse_beats_file(filepath):
    """Parse a .beats file to extract beat times and optionally beat numbers."""
    beats = []
    beat_numbers = []
    with open(filepath, 'r') as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) == 2:
                time, beat = float(parts[0]), int(parts[1])
                beat_numbers.append(beat)
            else:
                time = float(parts[0])
            beats.append(time)
    return beats, beat_numbers if beat_numbers else None

def infer_time_signature(beats, beat_numbers):
    """Determine the time signature(s) based on beat positions and numbers."""
    if beat_numbers:
        # Detect repeating bar lengths
        bar_lengths = np.diff([i for i, b in enumerate(beat_numbers) if b == 1])
        if len(bar_lengths) > 0:
            common_bar_length = Counter(bar_lengths).most_common(1)
            if common_bar_length:
                return f"{common_bar_length[0][0]}/4"
    
    # Fallback: Estimate from beat spacing if no downbeats are given
    if len(beats) < 2:
        return "Unknown"  # Not enough data to infer time signature
    
    diffs = np.diff(beats)
    avg_beat_interval = np.median(diffs) if len(diffs) > 0 else None
    
    if avg_beat_interval is None or avg_beat_interval == 0:
        return "Unknown"
    
    # Check periodic patterns in beat intervals
    intervals = np.round(diffs / avg_beat_interval).astype(int)
    
    if len(intervals) == 0:
        return "Unknown"

    common_pattern = Counter(intervals).most_common(1)
    if not common_pattern:
        return "Unknown"

    return f"{common_pattern[0][0]}/4"


def process_annotations(data_dir, skip_folders=[]):
    """Process all beat annotation files and return a dictionary of time signatures."""
    time_signatures = defaultdict(list)
    
    for root, dirs, files in os.walk(data_dir):
        if any(skip in root for skip in skip_folders):
            continue
        
        dataset = os.path.basename(root)
        for file in files:
            if file.endswith(".beats"):
                filepath = os.path.join(root, file)
                beats, beat_numbers = parse_beats_file(filepath)
                ts = infer_time_signature(beats, beat_numbers)
                # time_signatures[ts].append(dataset)

                # get just filename to save
                filename = os.path.basename(filepath)
                # print(filename)
                time_signatures[ts].append(filename)

    
    return time_signatures

def plot_time_signature_distribution_frequency(time_signatures, output_file='time_signature_distribution.pdf', plot_title="Time Signature Distribution"):
    """Plot a bar chart where each time signature has a fixed color across different datasets."""
    
    # Define a fixed color mapping for time signatures
    predefined_time_signatures = ["2/4", "3/4", "4/4", "5/4", "6/4"]
    color_map = {ts: color for ts, color in zip(predefined_time_signatures, plt.cm.Paired(np.linspace(0, 1, len(predefined_time_signatures))))}

    # Get counts of occurrences
    ts_counts = {ts: len(filepaths) for ts, filepaths in time_signatures.items()}

    # Ensure consistency by including all predefined time signatures (even if they don't exist in this dataset)
    all_time_sigs = predefined_time_signatures  # Use the predefined order
    counts = [ts_counts.get(ts, 0) for ts in all_time_sigs]  # Get counts, default to 0 if missing

    # Plot bar chart
    fig, ax = plt.subplots(figsize=(10, 6))
    bars = ax.bar(all_time_sigs, counts, color=[color_map[ts] for ts in all_time_sigs], width=0.6)

    # Labels and formatting
    ax.set_xlabel("Time Signature", fontsize=14)
    ax.set_ylabel("Number of Occurrences", fontsize=14)
    ax.set_title(plot_title, fontsize=16)
    ax.grid(axis="y", linestyle="--", linewidth=0.5)

    # Legend (only for the time signatures present in the dataset)
    legend_patches = [plt.Rectangle((0,0),1,1, color=color_map[ts]) for ts in all_time_sigs if ts in ts_counts]
    ax.legend(legend_patches, [ts for ts in all_time_sigs if ts in ts_counts], title="Time Signatures", fontsize=12)

    # Save and display
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.savefig(output_file, format='pdf', dpi=300, bbox_inches='tight')
    plt.show()


def print_percentages_of_timesignatures(time_signatures):
    # Calculate the total number of time signatures
    total_time_signatures = sum(len(v) for v in time_signatures.values())

    # Calculate the percentage of each time signature
    time_signature_percentages = {k: (len(v) / total_time_signatures) * 100 for k, v in time_signatures.items()}

    # Print the percentages
    for ts, percentage in time_signature_percentages.items():
        print(f"Time Signature: {ts}, Percentage: {percentage:.2f}%")


def save_time_signatures_to_file(time_signatures, filename="time_signatures.json"):
    """Save time signatures to a JSON file."""
    with open(filename, "w") as f:
        json.dump(time_signatures, f, indent=4)
```

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import numpy as np

def plot_time_signature_performance(json_path, csv_path, metric):
    """
    Plots a box plot of the specified metric per time signature.

    Parameters:
    - json_path: str, path to the JSON file containing time signature mappings.
    - csv_path: str, path to the CSV file with performance metrics.
    - metric: str, the metric to analyze (e.g., "F-measure_beat", "Cemgil_beat").
    """
    # Load JSON time signatures mapping
    with open(json_path, "r") as f:
        time_sig_dict = json.load(f)

    # Create a reverse mapping from .beats filenames to time signatures
    beats_to_time_sig = {filename: time_sig for time_sig, filenames in time_sig_dict.items() for filename in filenames}

    # Load CSV results
    df = pd.read_csv(csv_path)

    # Extract the .beats filename from the "Piece" column
    df["beats_filename"] = df["Piece"].apply(lambda x: re.search(r'gtzan_[^/]+_\d+', x).group(0) + ".beats" if re.search(r'gtzan_[^/]+_\d+', x) else None)

    # Map each row to a time signature using beats_to_time_sig
    df["Time Signature"] = df["beats_filename"].map(beats_to_time_sig)

    # Drop rows where the time signature is not found
    df = df.dropna(subset=["Time Signature"])

    # Define consistent colors for predefined time signatures
    predefined_time_signatures = ["2/4", "3/4", "4/4", "5/4", "6/4"]
    color_map = {ts: color for ts, color in zip(predefined_time_signatures, plt.cm.Paired(np.linspace(0, 1, len(predefined_time_signatures))))}

    # Plot box plot
    plt.figure(figsize=(10, 6))
    ax = sns.boxplot(x="Time Signature", y=metric, data=df, palette=color_map)
    
    # Title and labels
    plt.title(f"Performance per Time Signature ({metric})")
    plt.ylabel(metric)
    plt.xlabel("Time Signature")
    plt.xticks(rotation=45)
    plt.grid(True)

    # Show plot
    plt.show()
```

```python
def plot_weighted_time_signature_performance(json_path, csv_path, metric):
    """
    Plots a box plot of the specified metric per time signature with weighted means.
    
    Parameters:
    - json_path: str, path to the JSON file containing time signature mappings.
    - csv_path: str, path to the CSV file with performance metrics.
    - metric: str, the metric to analyze (e.g., "F-measure_beat", "Cemgil_beat").
    """
    # Load JSON time signatures mapping
    with open(json_path, "r") as f:
        time_sig_dict = json.load(f)
    
    # Create a reverse mapping from .beats filenames to time signatures
    beats_to_time_sig = {filename: time_sig for time_sig, filenames in time_sig_dict.items() for filename in filenames}
    
    # Load CSV results
    df = pd.read_csv(csv_path)
    
    # Extract the .beats filename from the "Piece" column
    df["beats_filename"] = df["Piece"].apply(lambda x: re.search(r'gtzan_[^/]+_\d+', x).group(0) + ".beats" if re.search(r'gtzan_[^/]+_\d+', x) else None)
    
    # Map each row to a time signature using beats_to_time_sig
    df["Time Signature"] = df["beats_filename"].map(beats_to_time_sig)
    
    # Drop rows where the time signature is not found
    df = df.dropna(subset=["Time Signature"])
    
    # Define predefined time signatures in the correct order
    predefined_time_signatures = ["4/4", "3/4", "2/4", "5/4"]  # Adjusted to match your original order
    
    # Filter data to only include predefined time signatures
    df = df[df["Time Signature"].isin(predefined_time_signatures)]
    
    # Convert Time Signature to categorical with the specific order
    df["Time Signature"] = pd.Categorical(
        df["Time Signature"], 
        categories=predefined_time_signatures, 
        ordered=True
    )
    
    # Define consistent colors for predefined time signatures
    color_map = {ts: color for ts, color in zip(predefined_time_signatures, plt.cm.Paired(np.linspace(0, 1, len(predefined_time_signatures))))}
    
    # Plot box plot
    plt.figure(figsize=(12, 7))
    ax = sns.boxplot(x="Time Signature", y=metric, data=df, palette=color_map, order=predefined_time_signatures)
    
    # Count instances per time signature
    time_sig_counts = df['Time Signature'].value_counts().reindex(predefined_time_signatures)
    
    # Calculate weighted means for each time signature
    weighted_means = {}
    total_samples = len(df)
    
    for ts in predefined_time_signatures:
        ts_data = df[df['Time Signature'] == ts][metric]
        n_samples = len(ts_data)
        weight = n_samples / total_samples  # Weight based on proportion of total samples
        weighted_means[ts] = np.mean(ts_data) * weight
    
    # Calculate overall weighted mean
    overall_weighted_mean = sum(weighted_means.values())
    
    # Add weighted means to the plot as markers
    for i, ts in enumerate(predefined_time_signatures):
        plt.plot(i, weighted_means[ts], marker='.', markersize=15, color='red', 
                 label='Weighted Mean' if i == 0 else "")
    
    # Add horizontal line for overall weighted mean
    plt.axhline(y=overall_weighted_mean, color='darkred', linestyle='--', 
                label='Overall Weighted Mean')
    
    # Replace x-tick labels with time signature and count
    time_sig_labels = [f"{ts}\n(n={time_sig_counts[ts]})" for ts in predefined_time_signatures]
    ax.set_xticklabels(time_sig_labels, fontsize=12)


    # Title and labels
    plt.title(f"Performance per Time Signature on Test Set ({metric})\nWeighted by Instance Count", fontsize=16)
    plt.ylabel(metric, fontsize=14)
    plt.xlabel("Time Signature (number of instances)", fontsize=14)
    plt.grid(True, alpha=0.3)
    
    # Add legend
    plt.legend(loc='best')
    
    # Add annotations with weight values (optional)
    for i, ts in enumerate(predefined_time_signatures):
        weight = time_sig_counts[ts] / total_samples * 100
        plt.annotate(f"{weight:.1f}%", 
                    xy=(i, df[df['Time Signature'] == ts][metric].max() + 0.01), 
                    xytext=(0, 5),
                    textcoords='offset points',
                    ha='center')
    
    # Adjust layout to prevent clipping
    plt.tight_layout()
    
    # save and Show plot
    plt.savefig("time-signature-plots/weighted-performance-test.pdf", format='pdf', dpi=300, bbox_inches='tight')
    plt.show()

    
    # Display weighted means in raw value format (not proportional)
    raw_weighted_means = {}
    for ts in predefined_time_signatures:
        ts_data = df[df['Time Signature'] == ts][metric]
        raw_weighted_means[ts] = np.mean(ts_data)
    
    print("Raw means by time signature:", raw_weighted_means)
    print("Overall weighted mean:", overall_weighted_mean)
    
    # Return weighted statistics for further analysis (optional)
    weighted_stats = {
        'weighted_means_by_ts': weighted_means,
        'raw_means_by_ts': raw_weighted_means,
        'overall_weighted_mean': overall_weighted_mean,
        'instance_counts': time_sig_counts.to_dict()
    }
    
    return weighted_stats
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Load the data
file1 = "../beat-this-results-on-test/file_metrics_log_finetune_timestretch_aug.csv"
file2 = "../beat-this-results-on-test/final-model-track-metrics.csv"

df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)

# Common column and metric
common_column = "Piece"
metric = "F-measure_beat"

# Merge all tracks
df_merged = df1[[common_column, metric]].merge(
    df2[[common_column, metric]], 
    on=common_column, 
    suffixes=("_finetune", "_original")
)

# Extract genre from 'Piece' (assuming structure like 'gtzan/gtzan_metal_00019/track.npy')
df_merged["Genre"] = df_merged[common_column].apply(lambda x: x.split('/')[1].split('_')[1] if '/' in x else "Unknown")

# Set up color palette
genres = df_merged["Genre"].unique()
palette = sns.color_palette("husl", len(genres))  # Use a distinct color for each genre
genre_colors = {genre: palette[i] for i, genre in enumerate(genres)}

# Scatter plot with genre-based coloring
plt.figure(figsize=(8, 8))
for genre, color in genre_colors.items():
    subset = df_merged[df_merged["Genre"] == genre]
    plt.scatter(
        subset[f"{metric}_original"], subset[f"{metric}_finetune"], 
        alpha=0.7, edgecolors="k", label=genre, color=color
    )

# Add a reference line (y = x)
plt.plot([0, 1], [0, 1], linestyle="--", color="red", label="No Change (y=x)")

# Labels and title
plt.xlabel("F1 Score (Original Model)")
plt.ylabel("F1 Score (Fine-tuned Model)")
plt.title("Comparison of F1 Scores by Genre")
plt.legend(title="Genre", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(True)
plt.xlim(0, 1)
plt.ylim(0, 1)

# Show the plot
plt.show()
```

```python
# Define which genres to plot
selected_genres = ["blues"]  # Modify this list to include more genres

# Filter the dataframe
df_filtered = df_merged[df_merged["Genre"].isin(selected_genres)]

# Count improvements, no change, and worsened scores
num_improved = (df_filtered[f"{metric}_finetune"] > df_filtered[f"{metric}_original"]).sum()
num_worse = (df_filtered[f"{metric}_finetune"] < df_filtered[f"{metric}_original"]).sum()
num_same = (df_filtered[f"{metric}_finetune"] == df_filtered[f"{metric}_original"]).sum()
num_total = len(df_filtered)

# Calculate percentages
percent_improved = (num_improved / num_total) * 100 if num_total > 0 else 0
percent_worse = (num_worse / num_total) * 100 if num_total > 0 else 0
percent_same = (num_same / num_total) * 100 if num_total > 0 else 0

# Scatter plot with genre-based coloring (only for selected genres)
plt.figure(figsize=(8, 8))
for genre in selected_genres:
    subset = df_filtered[df_filtered["Genre"] == genre]
    plt.scatter(
        subset[f"{metric}_original"], subset[f"{metric}_finetune"], 
        alpha=0.7, edgecolors="k", label=genre, color=genre_colors[genre]
    )

# Add a reference line (y = x)
plt.plot([0, 1], [0, 1], linestyle="--", color="red", label="No Change (y=x)")

# Labels and title
plt.xlabel("F-measure Score (Original Model)")
plt.ylabel("F-measure Score (Fine-tuned Model)")
plt.title(f"Comparison of F-measure Scores for {', '.join(selected_genres).title()} Genre(s)")
plt.legend(title="Genre", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(True)
plt.xlim(0, 1)
plt.ylim(0, 1)

plt.savefig("time-signature-plots/blues-comparison-f-measure-scores.pdf", format='pdf', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()

# Print improvement stats
print(f"Total tracks in selected genres: {num_total}")
print(f"Tracks improved after fine-tuning: {num_improved} ({percent_improved:.2f}%)")
print(f"Tracks stayed the same: {num_same} ({percent_same:.2f}%)")
print(f"Tracks worsened after fine-tuning: {num_worse} ({percent_worse:.2f}%)")
```
