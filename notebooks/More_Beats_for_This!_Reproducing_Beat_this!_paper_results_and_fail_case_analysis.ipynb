### estimate the time signature of worst tracks from the .beats files

```python
import numpy as np

def detect_time_signatures(beat_times, beat_numbers):
    """
    Detect time signatures from beat times and numbers.

    Args:
        beat_times: NumPy array of timestamps
        beat_numbers: NumPy array of beat numbers

    Returns:
        Dictionary mapping time signatures to their start and end timestamps
    """
    # Convert beat numbers to integers if they aren't already
    beat_numbers = beat_numbers.astype(int)

    # Find the maximum beat number which likely indicates the time signature numerator
    max_beat = int(np.max(beat_numbers))

    # Dictionary to store time signature changes
    time_signatures = {}

    # Check for consistency in the pattern
    def is_consistent_pattern():
        """Check if there's a consistent repeating pattern in the beat numbers"""
        # Find indices where the beat resets to 1 (likely indicates start of a measure)
        measure_starts = np.where((beat_numbers == 1) &
                                 (np.roll(beat_numbers, 1) > 1))[0]

        if len(measure_starts) < 2:
            return False, None

        # Check if the pattern length between measure starts is consistent
        pattern_lengths = np.diff(measure_starts)

        # If all patterns are the same length, we have a consistent time signature
        if np.allclose(pattern_lengths, pattern_lengths[0], rtol=0.1):
            return True, pattern_lengths[0]

        return False, pattern_lengths

    consistent, pattern_info = is_consistent_pattern()

    # Function to handle segments with different time signatures
    def analyze_segments():
        results = {}
        current_start_idx = 0
        current_sig = None

        # Find potential measure boundaries (where beat number decreases)
        potential_boundaries = np.where(np.diff(beat_numbers) < 0)[0] + 1

        if len(potential_boundaries) == 0:
            # No complete measures found, use highest beat as best guess
            results[f"{max_beat}/4"] = {
                "start": beat_times[0],
                "end": beat_times[-1]
            }
            return results

        # Check consistency within segments
        for i in range(len(potential_boundaries)):
            if i == 0:
                continue  # Skip the first boundary as we might start mid-measure

            segment_start = potential_boundaries[i-1]
            segment_end = potential_boundaries[i] if i < len(potential_boundaries) else len(beat_numbers)

            segment_beats = beat_numbers[segment_start:segment_end]

            # Determine time signature for this segment
            sig_numerator = int(np.max(segment_beats))
            sig = f"{sig_numerator}/4"

            if current_sig is None:
                current_sig = sig
                current_start_idx = segment_start
            elif sig != current_sig:
                # Time signature change detected
                results[current_sig] = {
                    "start": beat_times[current_start_idx],
                    "end": beat_times[segment_start - 1]
                }
                current_sig = sig
                current_start_idx = segment_start

        # Add the last segment
        if current_sig is not None:
            results[current_sig] = {
                "start": beat_times[current_start_idx],
                "end": beat_times[-1]
            }

        return results

    # For a consistent pattern, return a single time signature
    if consistent:
        time_signatures[f"{max_beat}/4"] = {
            "start": beat_times[0],
            "end": beat_times[-1]
        }
    else:
        # Try alternative approach for inconsistent or complex patterns
        time_signatures = analyze_segments()

        # If still no results, use a simple approach based on max beat number
        if not time_signatures:
            time_signatures[f"{max_beat}/4"] = {
                "start": beat_times[0],
                "end": beat_times[-1]
            }

    return time_signatures

def analyze_beats_file(beats_file):
    """
    Analyze time signature from a .beats file.

    Args:
        beats_file: Path to the .beats file

    Returns:
        Dictionary mapping time signatures to their start and end timestamps
    """
    # Load data directly using NumPy
    beat_times, beat_numbers = np.loadtxt(beats_file, unpack=True)
    return detect_time_signatures(beat_times, beat_numbers)
```

```python
import numpy as np
from collections import Counter

def detect_time_signature(beats_file):
    # Load beats data
    beat_times, beat_numbers = np.loadtxt(beats_file, unpack=True)

    # Find repeating cycle pattern
    diffs = np.diff(beat_numbers)  # Differences between consecutive beats
    cycle_lengths = []

    current_cycle = []
    for i in range(len(diffs)):
        current_cycle.append(beat_numbers[i])
        if diffs[i] < 0:  # Detect restart of the beat cycle
            cycle_lengths.append(len(current_cycle))
            current_cycle = []

    # Count occurrences of cycle lengths
    if cycle_lengths:
        most_common_cycle = Counter(cycle_lengths).most_common(1)[0][0]
        return f"Detected Time Signature: {most_common_cycle}/4"
    else:
        return "Unable to determine time signature."

# Usage example
beats_file = "/content/data/annotations/gtzan/annotations/beats/gtzan_blues_00000.beats"
print(detect_time_signature(beats_file))
```
